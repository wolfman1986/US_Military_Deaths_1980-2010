```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

df = pd.read_csv('C:/Users/ashto/OneDrive/Desktop/Mid Term Data Sets/Military_Deaths_Exploration/U.S. Military Deaths by cause 1980-2010.csv')

calendar_year = df['Calendar Year']
active_duty = df['Active Duty']

# Set the desired figure size
plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

plt.plot(calendar_year, active_duty)
plt.xlabel('Calendar Year')
plt.ylabel('Population')
plt.title('Active Duty Personnel')

# Customize x-axis tick marks and labels
plt.xticks(calendar_year, rotation=45)  # Set x-axis tick marks to the calendar year values

# Customize y-axis tick marks and labels
y_ticks = np.linspace(active_duty.min(), active_duty.max(), 6)  # Set desired number of y-axis tickers
plt.yticks(y_ticks)

def format_y_tick(x, pos):
    """Custom formatting function for y-axis tick labels"""
    millions = x / 1000000
    if millions.is_integer():
        return f'{int(millions)}M'
    else:
        return f'{millions:.1f}M'

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_y_tick))  # Format tick labels as '1M', '1.1M', '1.2M', etc.

plt.tight_layout()  # Adjust spacing to prevent overlapping

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('U.S. Military Deaths by cause 1980-2010.csv')
df.columns = df.columns.str.strip()

calendar_year = df['Calendar Year']
total_deaths = df['Total Deaths']
accident_deaths = df['Accident']

plt.figure(figsize=(10, 6))

# Create the bar graph for total deaths
plt.bar(calendar_year, total_deaths, label='Total Deaths')

# Create the line graph for accidents
plt.plot(calendar_year, accident_deaths, color='red', label='Accidents')

plt.xlabel('Calendar Year')
plt.ylabel('Number of Deaths')
plt.title('Comparison of Total Deaths and Accidents Over Calendar Years')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Read the data from the CSV file
df = pd.read_csv('U.S. Military Deaths by cause 1980-2010.csv')
df.columns = df.columns.str.strip()

calendar_year = df['Calendar Year']
columns_to_plot = ['Total Deaths', 'Accident', 'Hostile Action', 'Illness', 'Pending', 'Self-Inflicted', 'Terrorist Attack', 'Undetermined']

# Set the desired style for the plot using seaborn
sns.set(style='whitegrid')

plt.figure(figsize=(10, 6))

for column in columns_to_plot:
    plt.bar(calendar_year, df[column], label=column)

plt.xlabel('Calendar Year')
plt.ylabel('Number of Deaths')
plt.title('Comparison of Military Deaths by Cause Over Calendar Years')

# Set x-axis ticks to every year
plt.xticks(calendar_year, rotation=45)

# Set y-axis ticks in 100 increments
y_ticks = plt.yticks()[0]
plt.yticks(y_ticks, ['{:,.0f}'.format(tick) for tick in y_ticks])

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Assigned a full path to file and assigned to a dataframe
# df = pd.read_csv('C:/Users/ashto/OneDrive/Desktop/Mid Term Data Sets/Military_Deaths_Exploration/U.S. Military Deaths by cause 1980-2010.csv')
df = pd.read_csv('U.S. Military Deaths by cause 1980-2010.csv')

# Used the .strip() method to remove all leading and trailing spaces from string
df.columns = df.columns.str.strip()

# Assigned a variable for the x axis
calendar_year = df['Calendar Year']

# Assigned a variable for the y axis
self_inflicted_deaths = df['Self-Inflicted']


# Created a new figure object with a specified size
plt.figure(figsize=(10, 6))
# Created a bar chart, assigned calendar_year variable 
# to the x axis and self_inflicted_deaths to the y axis
plt.bar(calendar_year, self_inflicted_deaths)

# Labled the x axis
plt.xlabel('Calendar Years')

# Labeled the y axis
plt.ylabel('Number of Self-Inflicted Deaths')

# 
plt.title('Self-Inflicted Military Deaths Over Calendar Years')


# Set x-axis ticks to every year
plt.xticks(calendar_year, rotation=45)

# Set y-axis ticks in intervals of 10
y_ticks = range(0, self_inflicted_deaths.max() + 10, 10)
plt.yticks(y_ticks)

# Add markers for the significant events
plt.scatter(1987, self_inflicted_deaths.loc[df['Calendar Year'] == 1987].values[0], marker='*', color='orange', s=100, label='1987 Stock Market Crash')
plt.scatter(2001, self_inflicted_deaths.loc[df['Calendar Year'] == 2001].values[0], marker='*', color='red', s=100, label='Start of Afghanistan War')
plt.scatter(2008, self_inflicted_deaths.loc[df['Calendar Year'] == 2008].values[0], marker='*', color='green', s=100, label='2008 Financial Crisis')

# Add markers for the start of Iraq War, Desert Storm, and the surge in Iraq
plt.scatter(2003, self_inflicted_deaths.loc[df['Calendar Year'] == 2003].values[0], marker='*', color='blue', s=100, label='Start of Iraq War')
plt.scatter(1991, self_inflicted_deaths.loc[df['Calendar Year'] == 1991].values[0], marker='*', color='green', s=100, label='Start of Desert Storm (Gulf War)')
plt.scatter(2007, self_inflicted_deaths.loc[df['Calendar Year'] == 2007].values[0], marker='*', color='purple', s=100, label='Start of Surge in Iraq')

# Draw vertical lines for multiple markers on the top of the bar for 2001
plt.hlines(y=self_inflicted_deaths.loc[df['Calendar Year'] == 2001].values[0], xmin=2001, xmax=2001, colors=['purple', 'orange'], linestyles='solid')

plt.legend(loc='lower right', prop={'size': 8}, title='Significant Events')  # Move the legend to lower right and set smaller font size
plt.tight_layout()
plt.show()
```
