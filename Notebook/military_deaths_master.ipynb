```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('U.S. Military Deaths by cause 1980-2010.csv')
df.columns = df.columns.str.strip()

calendar_year = df['Calendar Year']
accident_deaths = df['Accident'].cumsum()
hostile_deaths = df['Hostile Action'].cumsum()
homicide_deaths = df['Homicide'].cumsum()
illness_deaths = df['Illness'].cumsum()
pending_deaths = df['Pending'].cumsum()
self_inflicted_deaths = df['Self-Inflicted'].cumsum()
terrorist_deaths = df['Terrorist Attack'].cumsum()
undetermined_deaths = df['Undetermined'].cumsum()

# Define a custom color palette
colors = ['#FF5733', '#3CB371', '#FFD700', '#4169E1', '#800080', '#FFA500', '#008080', '#FF00FF']

# Set the background color
plt.rcParams['axes.facecolor'] = '#F0F0F0'

plt.figure(figsize=(10, 6))

plt.plot(calendar_year, accident_deaths, marker='o', linestyle='-', color=colors[0], label='Accidents')
plt.plot(calendar_year, hostile_deaths, marker='o', linestyle='-', color=colors[1], label='Hostile')
plt.plot(calendar_year, homicide_deaths, marker='o', linestyle='-', color=colors[2], label='Homicide')
plt.plot(calendar_year, illness_deaths, marker='o', linestyle='-', color=colors[3], label='Illness')
plt.plot(calendar_year, pending_deaths, marker='o', linestyle='-', color=colors[4], label='Pending')
plt.plot(calendar_year, self_inflicted_deaths, marker='o', linestyle='-', color=colors[5], label='Self-Inflicted')
plt.plot(calendar_year, terrorist_deaths, marker='o', linestyle='-', color=colors[6], label='Terrorist Attack')
plt.plot(calendar_year, undetermined_deaths, marker='o', linestyle='-', color=colors[7], label='Undetermined')

plt.xlabel('Calendar Year')
plt.ylabel('Cumulative Number of Deaths')
plt.title('Cumulative Total Category Deaths 1980-2010')
plt.xticks(calendar_year, rotation=45)  # Set x-axis tick marks to the calendar year values
plt.legend()
plt.tight_layout()
plt.show()





```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import locale

# Set the locale for formatting numbers with commas
locale.setlocale(locale.LC_ALL, '')

# Read the data
df = pd.read_csv('U.S. Military Deaths by cause 1980-2010.csv')
df.columns = df.columns.str.strip()

# Define the years of interest
years = [1980, 1990, 2010]

# Define the color palette for the doughnut charts
colors = ['#333333', '#808080', '#90c7de']

# Iterate over the years to create doughnut charts
for year in years:
    # Filter the DataFrame for the current year
    df_year = df[df['Calendar Year'] == year]
    values = df_year[['Active Duty', 'Full-Time (est) Guard+Reserve', 'Selected Reserve FTE']].values.flatten()
    labels = ['Active Duty', 'Full-Time Guard+Reserve', 'Selected Reserve FTE']

    # Sort the values and labels in ascending order
    sorted_values, sorted_labels = zip(*sorted(zip(values, labels)))

    # Create the doughnut chart
    plt.figure(figsize=(6, 6))  # Adjust the figure size if needed

    # Create the outer ring (doughnut hole)
    plt.pie([1], colors=['white'], radius=0.65, startangle=90, wedgeprops=dict(width=0.3))

    # Create the inner pie chart with the customized color palette
    wedges, _ = plt.pie(sorted_values, labels=None, startangle=90, colors=colors, wedgeprops=dict(width=0.4))

    # Calculate the percentages and total number
    total = sum(sorted_values)
    percentages = [f'{v / total * 100:.1f}%' for v in sorted_values]

    # Format the values with commas
    formatted_values = [locale.format_string('%d', v, grouping=True) for v in sorted_values]

    # Create a single legend in the center of the doughnut
    legend_labels = [f'{label}\nTotal: {formatted_value}\nPercentage: {percentage}' for label, formatted_value, percentage in zip(sorted_labels, formatted_values, percentages)]
    plt.legend(wedges, legend_labels, loc='center', prop={'size': 8})

    # Equal aspect ratio ensures that the pie chart is circular
    plt.axis('equal')

    # Set the title font to Arial or Helvetica
    plt.suptitle(f'Distribution of Military Personnel in {year}', y=0.07, fontname='Arial')

    plt.show()






```
